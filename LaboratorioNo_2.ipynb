# Seleccion e indentificacion de la variable a a predecir considrando para el test el 20% de los datos.

```python
X = df.iloc[:, df.columns != 'Quantity']
y = df.loc[:, 'Quantity']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
X_test.shape[0]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
std_scaler = StandardScaler()

X_train = std_scaler.fit_transform(X_train)
X_test = std_scaler.transform(X_test)

X_train.shape, X_test.shape
```

# Aplicacion de red neuronal artificial ANN (artificial neural network)
## ANN No. 1

```python
from tensorflow.keras.optimizers import Adam
```

```python
entradas = X_train.shape[1]
lr = 0.05
batch_sz=50

regressor = Sequential()
regressor.add(Dense(input_dim=entradas, units=7, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=7, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=3, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=1, kernel_initializer='uniform', activation='relu'))
regressor.compile(optimizer=Adam(lr), loss='mean_squared_error', metrics=['mean_squared_error']) #regresion.
result1 = regressor.fit(X_train, y_train, batch_size=batch_sz, validation_split=0.2, verbose=1, epochs=50)

```

```python
result1.history
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
import time
```

```python
parametros = {
    'batch_size': [5, 15],
    'nb_epoch':[15, 20],
    'optimizer': ['adam', 'SGD']
}
```

```python
def reg_gs_v1(optimizer):
  regressor = Sequential()
  regressor.add(Dense(input_dim=entradas, units=3, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=8, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=2, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=1, kernel_initializer='uniform', activation='relu'))
  regressor.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mean_squared_error']) #regresion.
  result1 = regressor.fit(X_train, y_train, batch_size=batch_sz, validation_split=0.2, verbose=1, epochs=50)
  return regressor
```

```python
regresor_grid_search = KerasRegressor(build_fn=reg_gs_v1)
```

```python
grid_out = GridSearchCV(estimator=regresor_grid_search,
                        param_grid=parametros,
                        cv=10,
                        scoring='neg_mean_squared_error')
```

```python
train_gs_out = grid_out.fit(X_train, y_train, verbose=1)
```
